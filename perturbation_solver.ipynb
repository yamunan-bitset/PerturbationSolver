```python
import sympy as sp
from sympy.abc import n, i

power_of_epsilon = 7 # Number of iterations

print("Equation")
x = sp.Symbol("x")
LHS = x**5-x # USERINPUT
RHS = 1 # always a const; USERINPUT
quintic_expr = sp.Eq(LHS, RHS)
display(quintic_expr)

print("'Perturbed' Equation")
e = sp.Symbol("epsilon")
pLHS = x**5 - e*x # perturbed LHS; USERINPUT
perturbed_expr = sp.Eq(pLHS, RHS)
display(perturbed_expr)
```

```python
print("Set x to the following infinite series")
a = sp.IndexedBase(sp.Symbol("a"))
inf_series = sp.Sum(a[n]*e**n, (n, 0, power_of_epsilon))
coeffs_inf = {}
expanded_inf_series = inf_series.doit().expand()
sp.Eq(x, expanded_inf_series)
```

```python
print("Compute perturbed equation by plugging in the infinite series for x")
eqn = pLHS.subs(x, expanded_inf_series).expand()
sp.Eq(eqn, RHS)
```

```python
print("Equating all the like terms, and evaluating all the coefficients")

coeffs_inf = {}
eq = []
eq_s = []
for j in range(power_of_epsilon+1): # j in [0, 3]
    eq.append(sp.Eq(eqn.coeff(e, j), 1)) if j == 0 else eq.append(sp.Eq(eq_s[j-1], 0))
    coeffs_inf["a" + str(j)] = sp.solveset(eq[j], a[j], sp.Reals)
    tmp = eqn.coeff(e, j+1)
    for k in range(power_of_epsilon):
        try:
            tmp = tmp.subs(a[k], coeffs_inf["a" + str(k)].args[0])
        except KeyError:
            pass
    eq_s.append(tmp)

print(coeffs_inf)
```

```python
print("Finding the solution, by setting e = 1")

for j in range(power_of_epsilon+1):
    inf_series = inf_series.doit().subs(a[j], coeffs_inf["a" + str(j)].args[0])
inf_series = inf_series.subs(e, 1)
sol_poly = sp.Eq(x, inf_series.n(20))
display(sol_poly)
```
